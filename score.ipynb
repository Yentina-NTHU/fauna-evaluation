```python
scenarios = []
dialogs = []
scores = []
```

```python
import os
import re

path = './evaluator/sparky'

def extract_id(filename):
    match = re.search(r'scenario(\d+)\.txt', filename)
    return int(match.group(1)) if match else -1


scenario_files = [f for f in os.listdir(path) if f.startswith('scenario') and f.endswith('.txt')]
scenario_files.sort(key=extract_id)

for scenario_file in scenario_files:
    file_path = os.path.join(path, scenario_file)
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
        scenarios.append(content)

scenarios
```

```python
import os
import re
import pandas as pd

path = './report/Sparky/V0_0924_04'

def extract_id(filename):
    match = re.search(r'report_sparky_(\d+)\.csv', filename)
    return int(match.group(1)) if match else -1

dialog_files = [f for f in os.listdir(path) if f.startswith('report') and f.endswith('.csv')]
dialog_files.sort(key=extract_id)

for csv_file in dialog_files:
    file_path = os.path.join(path, csv_file)
    df = pd.read_csv(file_path)
    dialogs.append(df)

print(len(dialogs))
```

```python
for i in range(len(dialogs)):
    scenario = scenarios[i]
    print(f'Scenaroio {i+1}')
    print(scenario)
    print('. ' * 40)

    dialog = dialogs[i]
    for index, row in dialog.iterrows():
        print('User:  ', row['person_say'])
        print('Agent: ', row['animal_say'])

    score = input('Enter the score: ')
    scores.append(score)
    
    print('-' * 80)
```
